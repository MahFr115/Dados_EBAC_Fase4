#### O Bagging

O Bagging é uma ferramenta para desenvolvimento de modelos de predição, melhorados, de uma série de dados. A ideia ao se desenvolver o raciocício remete a algo como: dividir para melhorar.<br>
Separando para treinar as partes e assim, ter a modelagem, treino e teste em bases distintas, reber os resultados gerados pelo modelo e garantir que o resultado mais provável (o que estiver mais presentes nas repostas)  .  

#### Passo a passo
Para desenvolvimento de um Bagging precisamos:<br>
  1. Receber e tratar os dados da base<br> 
  2. Separar a base de dados dada em partes de igual tamanho, podendo repetir as linhas de dados e mudar a ordem de aparição.<br>
  3. Desenvolver o modelo que se qualifique melhor para essas bases.<br>
  4. Aplicar o modelo para previsões.<br>
  5. Receber os resultados dessas previsões.<br>
  6. Escolher entre os resultados previstos aquele que ter maior probabilidade de ocorrer. Em caso de resultados categorizados escolher o resultado mais frequênte, em caso de resultados específicos a média entre os obtidos.

#### Exemplos
Podemos exemplificar a utilização do Bagging com os seguintes itens: <br>
  - Tirar bolinhas, de 5 cores diferentes, de dentro de uma caixa, no escuro, as repondo a medida que as tira. Repetindo a ação 100 vezes. Para prever qual a próxima bolinha será retirada.<br>

Em uma caixa contendo bolinhas de cinco cores diferentes, retiro uma bolinha por vez, anotando aleatoriamente a cor sorteada e, em seguida, devolvendo-a à caixa. Repito esse processo várias vezes para gerar diferentes bases de dados. <br>
Com essas bases, construo modelos com o objetivo de analisar e prever o comportamento e a frequência das cores que aparecem.<br>
A partir das previsões geradas por cada modelo sobre a próxima cor provável, reúno os resultados em uma tabela. Nela, registro as opções e suas respectivas frequências de ocorrência, tomando como resposta final aquela que aparecer com maior frequência — ou seja, a maioria das previsões.


  - Jogar um dado normal 80 vezes. Para prever o próximo resultado da jogada<br>
 <br>

Em uma caixa com vários dados de seis lados, retiro um dado por vez, lanço-o e anoto o número que saiu, devolvendo o dado à caixa após cada jogada. Repito esse processo diversas vezes, criando diferentes bases de dados com os resultados registrados de forma aleatória.<br>
A partir dessas bases, construo diferentes modelos com o objetivo de prever qual número tende a sair com mais frequência nas próximas jogadas.<br>
Com os resultados obtidos, reúno as previsões em uma tabela, listando os números indicados e quantas vezes cada um foi previsto. A resposta final será o número mais frequentemente sugerido entre os modelos — ou seja, o voto da maioria.


  - Previsão de quantos produtos devem ser comprados por uma loja considerando as vendas nos períodos anteriores<br>
  <br>

Com base no histórico de vendas de um determinado produto, crio várias amostras aleatórias dos dados disponíveis, permitindo repetições (bootstrapping). <br>
Para cada amostra, desenvolvo um modelo de previsão que estima quantas unidades do produto devem ser compradas para o estoque futuro.<br>
Após gerar essas previsões com diferentes modelos, reúno os resultados e calculo a média das quantidades sugeridas. Essa média representa a estimativa final mais robusta, pois combina os aprendizados de múltiplas versões do modelo, reduzindo a influência de variações ou ruídos em uma única amostra.
